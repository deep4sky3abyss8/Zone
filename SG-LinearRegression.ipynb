IV. now, must implement linear regression algorithme & gradient disent ...

the h :  t0*1 + t1*x1 + ... + tn*xn

so :  
- t[ t0,t1,t2,...,tn ] 
- x[ 1,x1,x2,...,xn ]


````python
def h( t,x ) : 
    return t[0] + t[1]*x[1] + t[2]*x[2]
#---------------------------------------------------------------
def squerError( t, features_vestor, true_value ):
    
    N = len(features_vestor)
    sum_ = 0
    for i in range(N) : 
        sum_ += (h(t,features_vestor[i])-true_value[i])**2/N
    
    return sum_
#---------------------------------------------------------------
def update_t( old_t, features_vestor, true_value, alpha ): 
    
    N = len(features_vestor)
    sum_ = [ 0, 0, 0 ]
    
    for i in range(N) :
        
        absError = h(old_t, features_vestor[i]) - true_value[i]
        
        sum_[0] += absError*features_vestor[i][0]
        sum_[1] += absError*features_vestor[i][1]
        sum_[2] += absError*features_vestor[i][2]
        
    new_t = [ 0, 0, 0 ]
    for i in range(3): 
        new_t[i] = old_t[i] - 2*alpha*sum_[i]/N
    
    return new_t
#---------------------------------------------------------------
def train( t, features_vestor, true_value, alpha, max_iterations ) :
    
    for i in range(max_iterations) :
        
        t = update_t(t, features_vestor, true_value, alpha)
        if i%900 == 0 : print(f"iteration<{i}> SquerError<{squerError(t,features_vestor,true_value)}>")
    
    return t
#---------------------------------------------------------------
T = [ 3,1,1 ] # t0->0 t1->0 t2->0
T = train( T, X, Y, 0.001, 15000 )
print(T)
````
